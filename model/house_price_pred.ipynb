### Independent and Dependent features

```python
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
```

```python
X
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle 
pickle.dump(scaler, open('scaler.pkl', 'wb')) 
```

```python
X_train
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,Y_train)
```

### Print the coefficients and intercepts

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

### Prediction

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

### plot a scatter plot for the prediction

```python
plt.scatter(Y_test,reg_pred)
```

```python
residuals = Y_test - reg_pred
```

```python
residuals
```

### plotting residuals

```python
sns.displot(residuals,kind="kde")
```

```python
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(f" Mean squared error : {mean_squared_error(Y_test,reg_pred)}")
print(f"Mean absolute error : {mean_absolute_error(Y_test,reg_pred)}")
print(f"Root mean squared error : {np.sqrt(mean_squared_error(Y_test,reg_pred))}")
```

### R square and adjusted R-square

```python
from sklearn.metrics import r2_score
score = r2_score(Y_test,reg_pred)
print(f"R square : {score}")
```

```python
adjusted_r_score = 1 - (1-score)*(len(Y_test)-1)/(len(Y_test)-X_test.shape[1]-1)
print(f"Adjusted R square : {adjusted_r_score}")
```
